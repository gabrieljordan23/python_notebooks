# Desafio: utilizando parâmetros do json_normalize


Ao normalizar o objeto JSON durante a aula passamos simplesmente pd.json_normalize(<dados>) mas o método json_normalize possui diversos parâmetros para trabalhar com estruturas de dados mais complexas.

Você recebeu a tarefa de normalizar o seguinte arquivo JSON chamado “informacoes.json” referente a algumas informações de identificação de um cliente. O arquivo possui o seguinte conteúdo:



```
# {
  "nome": "João",
  "idade": 28,
  "enderecos": [
    {
      "tipo": "casa",
      "rua": "Rua A",
      "numero": 123,
      "cidade": "São Paulo"
    },
    {
      "tipo": "trabalho",
      "rua": "Rua B",
      "numero": 456,
      "cidade": "Rio de Janeiro"
    }
  ]
}
```



O resultado final deve ser um Pandas DataFrame.

````python
import pandas as pd

data = {
  "nome": "João",
  "idade": 28,
  "enderecos": [
    {
      "tipo": "casa",
      "rua": "Rua A",
      "numero": 123,
      "cidade": "São Paulo"
    },
    {
      "tipo": "trabalho",
      "rua": "Rua B",
      "numero": 456,
      "cidade": "Rio de Janeiro"
    }
  ]
}
''' A função pd.json_normalize() é usada para normalizar os dados em uma tabela.
Nesse caso, data é passado como o objeto JSON a ser normalizado. O parâmetro record_path
é definido como 'enderecos' para indicar que os dados das pessoas devem ser normalizados.
O parâmetro meta é definido como "nome" e "idade" para inserir a informação de nome e idade
em cada linha da tabela. Finalmente, o parâmetro errors é definido como 'ignore' para ignorar
quaisquer erros que possam ocorrer durante a normalização.'''

df = pd.json_normalize(data, record_path='enderecos', meta = ['nome', 'idade'], errors='ignore')
df
````

# Desafio: tratando os dados

```
{
  "pessoas": [
    {
      "nome": "João",
      "idade": "25",
      "endereco": {
        "rua": "Rua A",
        "numero": 123,
        "cidade": "São Paulo"
      },
      "telefones": [
        "11 1111-1111",
        "11 2222-2222"
      ]
    },
    {
      "nome": "Maria",
      "idade": 30,
      "endereco": {
        "rua": "",
        "numero": 456,
        "cidade": "Rio de Janeiro"
      },
      "telefones": [
        "21 3333-3333"
      ]
    }
  ]
}
```

Sua tarefa é realizar alguns tratamentos e deixar esse arquivo JSON como um Pandas DataFrame, sem dados vazios e com a tipagem das colunas de forma correta. Como isso pode ser feito?

````python
''''mport json

with open('dados.json', 'r') as f:
    dados = json.load(f)
'''
import pandas as pd

data = {
  "pessoas": [
    {
      "nome": "João",
      "idade": "25",
      "endereco": {
        "rua": "Rua A",
        "numero": 123,
        "cidade": "São Paulo"
      },
      "telefones": [
        "11 1111-1111",
        "11 2222-2222"
      ]
    },
    {
      "nome": "Maria",
      "idade": 30,
      "endereco": {
        "rua": "",
        "numero": 456,
        "cidade": "Rio de Janeiro"
      },
      "telefones": [
        "21 3333-3333"
      ]
    }
  ]
}

df = pd.json_normalize(data, record_path=['pessoas'])

df.info()

df['idade'] = df['idade'].astype(int)

df_filtrado = df.query('`endereco.rua` != ""')

df_filtrado = df[df['endereco.rua'] != ""]

````
