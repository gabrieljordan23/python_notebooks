# Desafio: utilizando parmetros do json_normalize


Ao normalizar o objeto JSON durante a aula passamos simplesmente pd.json_normalize(<dados>) mas o mtodo json_normalize possui diversos parmetros para trabalhar com estruturas de dados mais complexas.

Voc recebeu a tarefa de normalizar o seguinte arquivo JSON chamado informacoes.json referente a algumas informaes de identificao de um cliente. O arquivo possui o seguinte contedo:



```
# {
  "nome": "Joo",
  "idade": 28,
  "enderecos": [
    {
      "tipo": "casa",
      "rua": "Rua A",
      "numero": 123,
      "cidade": "So Paulo"
    },
    {
      "tipo": "trabalho",
      "rua": "Rua B",
      "numero": 456,
      "cidade": "Rio de Janeiro"
    }
  ]
}
```



O resultado final deve ser um Pandas DataFrame.

````python
import pandas as pd

data = {
  "nome": "Joo",
  "idade": 28,
  "enderecos": [
    {
      "tipo": "casa",
      "rua": "Rua A",
      "numero": 123,
      "cidade": "So Paulo"
    },
    {
      "tipo": "trabalho",
      "rua": "Rua B",
      "numero": 456,
      "cidade": "Rio de Janeiro"
    }
  ]
}
''' A funo pd.json_normalize()  usada para normalizar os dados em uma tabela.
Nesse caso, data  passado como o objeto JSON a ser normalizado. O parmetro record_path
 definido como 'enderecos' para indicar que os dados das pessoas devem ser normalizados.
O parmetro meta  definido como "nome" e "idade" para inserir a informao de nome e idade
em cada linha da tabela. Finalmente, o parmetro errors  definido como 'ignore' para ignorar
quaisquer erros que possam ocorrer durante a normalizao.'''

df = pd.json_normalize(data, record_path='enderecos', meta = ['nome', 'idade'], errors='ignore')
df
````


Sua tarefa  realizar alguns tratamentos e deixar esse arquivo JSON como um Pandas DataFrame, sem dados vazios e com a tipagem das colunas de forma correta. Como isso pode ser feito?

````python
''''mport json

with open('dados.json', 'r') as f:
    dados = json.load(f)
'''
import pandas as pd

data = {
  "pessoas": [
    {
      "nome": "Joo",
      "idade": "25",
      "endereco": {
        "rua": "Rua A",
        "numero": 123,
        "cidade": "So Paulo"
      },
      "telefones": [
        "11 1111-1111",
        "11 2222-2222"
      ]
    },
    {
      "nome": "Maria",
      "idade": 30,
      "endereco": {
        "rua": "",
        "numero": 456,
        "cidade": "Rio de Janeiro"
      },
      "telefones": [
        "21 3333-3333"
      ]
    }
  ]
}

df = pd.json_normalize(data, record_path=['pessoas'])

df.info()

df['idade'] = df['idade'].astype(int)

df_filtrado = df.query('`endereco.rua` != ""')

df_filtrado = df[df['endereco.rua'] != ""]

````
