```
#Q1 - lista 6 
import pandas as pd
import math
import numpy as np

df = pd.read_csv("https://www.dropbox.com/s/wias1rm7dt225b6/pontos-plano-18.csv?dl=1", sep = ",")

indice_p1 = int (input())
indice_p2 = int (input())
indice_p3 = int (input())

def calcula_distancia_euclidiana(df, indice_p1, indice_p2, indice_p3):

  x_1, y_1 = df[['X','Y']].loc[indice_p1]
  x_2, y_2 = df[['X','Y']].loc[indice_p2]
  x_3, y_3 = df[['X','Y']].loc[indice_p3]

  dis1 = ((x_2 - x_1)**2 + (y_2 - y_1)**2)**(1/2)
  dis2 = ((x_3 - x_2)**2 + (y_3 - y_2)**2)**(1/2)
  dis3 = ((x_1 - x_3)**2 + (y_1 - y_3)**2)**(1/2)
  sumDis = dis1 + dis2 + dis3
  return "%.2f" %sumDis

print(calcula_distancia_euclidiana(df, indice_p1, indice_p2, indice_p3))



```

```
#Q3 - lista 6 

import pandas as pd
#import math

df = pd.read_csv("https://www.dropbox.com/s/8tq1bxi8acps36p/latitude-longitude-bairros.csv?dl=1", sep = ";")

indice_city1 = int(input())
indice_city2 = int(input())
#latidude1, longitude1, latidude2, longitude2 = def calcula_distancia_km(df, indice_city1, indice_city2)

def distancia_terra(lat1, lon1, lat2, lon2):
  from math import sin, cos, sqrt, atan2, radians
  lat1, lon1, lat2, lon2 = [radians(lat1), radians(lon1), radians(lat2), radians(lon2)]
  dlon, dlat = [lon2 - lon1,  lat2 - lat1]
  a = sin(dlat / 2) ** 2 + cos(lat1) * cos(lat2) * sin(dlon / 2) ** 2
  return 6371.0 * 2 * atan2(sqrt(a), sqrt(1 - a))   # 6371.0 = raio aproximado da terra

def calcula_distancia_km(df, indice_city1, indice_city2):

	latidude1, longitude1 = df[['latitude1','longitude']].loc[indice_city1]
	latidude2, longitude2 = df[['latitude','longitude']].loc[indice_city2]
  cal = distancia_terra(latidude1, longitude1, latidude2, longitude2)
	return cal

#latidude1, longitude1, latidude2, longitude2 = def calcula_distancia_km(df, indice_city1, indice_city2)
resposta = calcula_distancia_km(df, indice_city1, indice_city2)
print("%.2f" %resposta)
```

```
#Q1 - lista 7 
import math
import pandas as pd
import numpy as np

df = pd.read_csv("https://www.dropbox.com/s/qgp1gqxr468qj7u/fake-classrooms10.csv?dl=1", sep = ",")

#recebendo dados via input
notaA = float(input())
notaB = float(input())
notaC = float(input())
notaD = float(input())
#fução que calcula o conceito
def calculaConceito(df, notaA, notaB, notaC, notaD):
	media = df['Nota']
	if 10 >= media >= notaA:
		return 'A'
	elif notaA >= media >= notaB:
		return 'B'
	elif notaB >= media >= notaC:
		return 'C'
	elif notaC >= media >= notaD:
		return 'D'
	else:
		return 'F'
#função que formata o conceito calculado anteriormente
def formataConceito(df):
	hist = df['Conceito'].value_counts()
	resp = ''
	if 'A' in hist.keys(): # verifica se tem algum conceito A
		resp += 'A: %d\n' % hist['A']
	if 'B' in hist.keys():
		resp += 'B: %d\n' % hist['B']
	if 'C' in hist.keys():
		resp += 'C: %d\n' % hist['C']
	if 'D' in hist.keys():
		resp += 'D: %d\n' % hist['D']
	if 'F' in hist.keys():
		resp += 'F: %d\n' % hist['F']
	return str(resp)


# função para calcular a m ́edia em cada linha da tabela
def calculaNota(linha):
	return (linha["Prova 1"] + linha["Prova 2"] + linha["Trabalho"]) / 3
# criar a coluna ’Notas’ e utilizar o m ́etodo apply, com argumentos:
# a fun ̧c~ao criada e axis=1 (para processar cada linha de df)
df['Nota'] = df.apply(calculaNota, axis=1)
df['Conceito'] = df.apply(calculaConceito, axis = 1, args = [notaA, notaB, notaC, notaD])

print(formataConceito(df))
```

```
#Q2 - lista 8

import pandas as pd

mercados = pd.read_csv("https://www.dropbox.com/s/7vpjxlnv6555ls9/fake-mercado28.csv?dl=1"), 
pd.read_csv("https://www.dropbox.com/s/yr1uby5ylaz9h7j/fake-mercado18.csv?dl=1"), 
pd.read_csv("https://www.dropbox.com/s/tcbxw0s0edp3ghm/fake-mercado14.csv?dl=1")

lista = []

for x in range(0,10):
  produto = input()
  itens_carrinho = []
  for y in range(0,3):
    filtro = mercados[y].query("Item == '"+produto+"'")
    filtro2 = filtro.values.tolist()
    itens_carrinho.append(filtro2)
  lista.append(itens_carrinho)

lista2 = []

for itens_carrinho in lista:
  valores = []
  for mercados in itens_carrinho:
    for colunas in mercados:
      valores.append(colunas[2])
  menor = min(valores)

  for z in range(0,3):
    if (valores[z] == menor):
      mercados = valores.index(menor)+1
      lista2.append([colunas[0], colunas[1], menor, mercados])

mercados_novo = pd.DataFrame(lista2, columns=["Categoria", "Item", "Menor Valor", "Mercado"])
mercados_novo.sort_values(by=["Categoria", "Item"], inplace=True)
print(mercados_novo.to_string(index = False))
```

```
#Q5 - lista 9
import random as rd
#declarando as varáveis com valores recebidos via teclado
semente = int(input())
rd.seed(semente)
saldo = int(input())
aposta = int(input())
numero = int(input())
#declarando variáveis internas
prosperou = 0
faliu = 0
total = 0
#loop que calcula os ganhos e perdas do jogador
for i in range(0, numero):
	total = saldo
	while total >= 0 and total <= (2*saldo):
		numero = rd.randint(4,5)
		if numero == 4:
			total = total + aposta
		else:
			total = total - aposta
	if total <= 0:
		faliu = faliu + 1
	else:
		prosperou = prosperou + 1
#saídas na tela
print("O jogador faliu", faliu, "vezes")
print("O jogador prosperou", prosperou,"vezes")
#condicionais para status de ganhos e perdas do jogador
if faliu < prosperou:
	print("Numero de vezes que o jogador faliu eh menor que o numero de vezes que prosperou")
elif faliu == prosperou:
	print("Numero de vezes que o jogador faliu eh igual ao numero de vezes que prosperou")
else: 
	print("Numero de vezes que o jogador faliu eh maior que o numero de vezes que prosperou")
```
